## **1. コンピュータの判断方法**

### **1.1 コンピュータには「目」がない！？👀**

コンピュータは人間みたいに「目」や「感覚」はありません。  
でも、**ものの色、大きさ、形**などを**数字**で表せば、それを見て考えることができるんだ！

### **1.2 たとえば、りんごとバナナを見分けるとき**

| 色（0〜1） | 大きさ（cm） | 形（まるさ） | 果物の名前 |
|------------|--------------|----------------|--------------|
| 0.8        | 7            | 0.9            | りんご       |
| 0.3        | 12           | 0.2            | バナナ       |
| 0.85       | 6.5          | 0.95           | りんご       |

> ✅「色」や「形」を**数字で表す**ことで、コンピュータにも違いがわかるようになるんだ！

### **1.3 データが多いほど、もっと上手になる！**

- 1つか2つだけのデータでは、判断がむずかしい。
- でも、たくさんの例を見れば見るほど、「パターン」を見つけて上手に分けられるようになる！

たとえば…

- りんご：100個見た → だいたいの特徴がわかってきた！
- バナナ：5個しか見てない → 間違えやすいかも…

## **2. 間違えることもある！？😱**

### **2.1 間違ったらダメ？→ いいえ、学びのチャンス！**

コンピュータも最初は間違えることがあります。

たとえば：
- 緑色のりんご → 「バナナかも？」と間違える
- 小さいバナナ → 「りんごっぽい！」と間違える

でも大丈夫！間違えたあとに「正しい答え」を教えてあげれば、またひとつ賢くなるんです！

### **2.2 間違いを直すには？**

間違えたときの対処法は2つ！

1. **もっとたくさんのデータを見せる**  
　→ 色んなパターン（例：緑のりんご、ちっちゃいバナナ）を知ってもらう。

2. **特徴を工夫する**  
　→ 「におい」「重さ」など、**他の特徴**も使って判断できるようにする。

> まるで、知らない友達の名前を覚えるときに「背が高い」「声が大きい」などの特徴をたくさん覚えるみたいな感じ！

## **3. ちょっと復習 – 分類の流れ**

| ステップ | やること |
|----------|----------|
| 1. データを集める | 例：りんごとバナナの写真と特徴 |
| 2. 正しい答えをつける | 「これはりんご」「これはバナナ」など |
| 3. 学習する | コンピュータがルールを見つける |
| 4. 予測してみる | 新しい果物が出てきたら判断してみる |
| 5. 間違えたらやり直す | もっと良いルールを探す！ |

## **4. Pythonで「間違いと学び直し」をやってみよう！**

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# データを読み込み
iris = load_iris()
X = iris.data
y = iris.target

# データを分ける（80% 学習用、20% テスト用）
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# モデルを作って学習
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X_train, y_train)

# テストデータで予測
y_pred = model.predict(X_test)

# 正しく当てた割合（正解率）を見る
score = accuracy_score(y_test, y_pred)
print("正解率:", score)
```

> ✅ もし正解率があまり高くなかったら…
> - データをもっと増やす
> - モデルのやり方（kの数）を変えてみる
> → そうやって少しずつ「学び直し」していきます！

## **5. アクティビティ：フルーツ仕分けゲーム**

### 目的
- 特徴をもとに「分ける」ことを体験する
- 間違えたときに「どんな特徴が足りなかったか？」を考える

### やり方

1. **フルーツのカードを準備しよう！**
　→ イラストまたは写真付きで、「色」「大きさ」「形」などの特徴を書いたカードを使う

2. **みんなで分けてみよう！**
　→ これは「りんご」かな？それとも「もも」かな？理由も一緒に発表！

3. **先生が正解を発表！**
　→ 間違えてもOK！「なんで間違えたのか」を話し合おう

4. **新しい特徴を考えてみよう！**
　→ たとえば「におい」「手ざわり」など、他に使える情報を探してみよう！

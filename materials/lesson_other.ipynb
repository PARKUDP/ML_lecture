# **実践編 – ぼくの・わたしのAIチャレンジ！**

## **目標：AIの力を使って、生活の中の“困った”を解決しよう！**

## **全体構成（全4回）**

| 回 | タイトル | 内容 |
|----|----------|------|
| 1 | AIでなんとかしたい！を考える | 身近な問題を見つけて、AIで解決できるか考える |
| 2 | どの学習を使えばいい？ | 分類・回帰・強化学習のどれが合うかを選ぶ |
| 3 | 簡単なデータやルールを作ってみよう | 手作りのデータやルールを考えて、自分で“学習体験” |
| 4 | 発表＆フィードバック | 自分の考えたAIを紹介して、クラスみんなでコメント！

## **第1回：AIでなんとかしたい！を考える**

### ゴール：
- 「困ってること」や「もっと便利にしたいこと」を見つけよう！

### アイデア例：
- 忘れ物が多い → 「チェックリストAI」
- 宿題がたまりがち → 「計画表AI」
- 勉強中に眠くなる → 「ごほうびAI（強化学習）」

### ワーク：
- ふだんの生活で「困っていること」を紙に書こう
- それをAIでどうやって助けられるか考えてみよう！

## **第2回：どの学習方法がぴったり？**

### ゴール：
- 自分のアイデアに合う「学習方法」を考えてみよう！

| 学習の種類 | こんなときに使える |
|------------|-------------------|
| 分類       | 「これは何？」を知りたいとき（例：動物の名前） |
| 回帰       | 「どれくらい？」を予測したいとき（例：宿題が終わる時間） |
| 強化学習   | 「どう動けばいい？」を考えたいとき（例：ごほうびのタイミング） |

### ワーク：
- 自分のアイデアに合う学習方法にチェック 
- なぜそれが合ってると思ったかも書いてみよう！

## **第3回：手作りデータ＆シミュレーションに挑戦！**

### ゴール：
- 自分で小さな「学習ゲーム」や「表」を作って、AI体験してみよう！

### 例1：分類
- おやつカード（チョコ・キャンディ・ガム）を作って、特徴（色・形）で分類してみる

### 例2：回帰
- 「曜日と宿題の量」→ 明日の量を予測してみよう！

| 曜日 | 宿題の量（ページ） |
|------|-------------------|
| 月   | 2                 |
| 火   | 3                 |
| 水   | 4                 |
| 木   | ?（予測！）        |

### 例3：強化学習
- ボールを投げてごほうびシールをもらう → 一番点が高くなる方法を見つける！

## **第4回：みんなに発表しよう！**

### ゴール：
- 自分が考えた「ぼくの・わたしのAI」を紹介してみよう！

### 発表ポイント：
1. どんな問題を解決したい？
2. どんな学習方法を使った？
3. どんな工夫をした？
4. やってみてどうだった？むずかしかった？たのしかった？

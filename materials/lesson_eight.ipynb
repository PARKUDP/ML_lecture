# **第8回 – 振り返り**  
## **1. これまでどんなことを学んだ？**

### **1.1 機械学習ってなに？（第1回）**

- コンピュータが**たくさんのデータ**から**自分でルールを学ぶ技術**
- 人間がルールをぜんぶ書かなくても、データから考えてくれる！

### **1.2 分類（ぶんるい）って？（第2・3回）**

- **「これは何？」を当てる**学び方  
　例）猫と犬を見分ける、お菓子を分ける

- 特徴（とくちょう）を使ってグループに分ける
- お菓子わけゲームや、花の分類をやったね！

### **1.3 回帰（かいき）って？（第4・5回）**

- **「どれくらい？」を予測する**学び方  
　例）明日の気温、身長、おこづかいの増え方

- 平均を見たり、数字の変化を考えたりしたね！
- サイコロの目の予測ゲームもやったね

### **1.4 強化学習（きょうかがくしゅう）って？（第6・7回）**

- **やってみて、失敗しながら覚える学び方！**  
　例）ゲームで敵を倒す、迷路をゴールする

- 報酬（ごほうび）がもらえる動きが何かを学ぶ！
- 迷路やボール投げゲーム、楽しかったね！

## **2. まとめて比べてみよう！**

| 回帰         | 数字を予測する     | 気温や身長の予測     |
| 強化学習     | 試してうまくなる   | ゲームで勝つ、迷路を抜ける |

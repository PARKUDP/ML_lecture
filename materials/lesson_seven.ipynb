## **1. くり返せば上手くなる！**

### **1.1 人間もAIも、最初はうまくいかない！**

ピアノも、けん玉も、自転車も……  
最初からうまくできる人なんていないよね？

- 何度も練習して、少しずつコツがわかってくる  
- 最初は失敗しても、**「次はこうしよう」**と考えて、どんどん成長する！

### それって、まさに強化学習！

強化学習も、「トライ（ためす） → 失敗 → 学ぶ → もう一度トライ」のくり返しです。

## **2. 失敗から学ぶしくみ**

### **2.1 「失敗」って悪いこと？**

ぜんぜんそんなことない！  
むしろ、**失敗するからこそ学べる！**

たとえば…

| 行動           | 結果            | 報酬（ごほうび） |
|----------------|-----------------|------------------|
| 適当にボールを投げた | ゴールから外れた     | 0点               |
| 力をこめて投げた     | ゴールの上を通りすぎた | 0点               |
| ちょっと弱く投げた   | ゴールに入った！       | 10点              |

→ 「どうやったらゴールに入るのか？」を**失敗から見つける**のがポイント！

### **2.2 少しずつルールを覚えていく**

1回目 → はずれる  
2回目 →ちょっと近づく  
3回目 → 入った！  
4回目 → もう一度入った！

→ 何度もやって、**「うまくいくやり方」を自分で見つける**んだ！

## **3. AIはどうやって「うまくなる」の？**

### AIは、試行（ためす）と報酬（点数）をセットで覚えていきます

- **どんな行動をしたか？**
- **そのとき、何点もらえたか？**
- → たくさんためて、「どの行動がいちばんいいか」を少しずつ学んでいく！

## **4. アクティビティ：ボールをゴールに入れるゲーム**

### ゲームの目的：
- くり返しやって、少しずつ上手になる体験をする
- 強化学習の「試行錯誤（しこうさくご）」を体で感じる！

### 準備するもの：

- 小さめのボール
- ゴール（かご、箱、紙の円など）
- 得点カード（正解なら10点！）

### ゲームの流れ：

1. **スタート地点を決めて、ボールを投げる**
　→ 1人1回ずつ投げてみよう！

2. **入らなかったら、なぜか考える**
　→ 強すぎた？角度は？場所を変えたら？

3. **くり返して投げてみる**
　→ だんだん入るようになるはず！

4. **いちばん上手になった人は？**
　→ 「どこがよかったか」発表してみよう！

### ポイント！

- 1回目で入らなくても気にしない！
- 2回目、3回目と「うまくいくやり方」を少しずつ覚えていく
- → これがまさに「強化学習の考え方」なんだ！

## **5. ちょっとだけPythonでもやってみよう！（参考）**

```python
import random

score = 0

for i in range(10):
    strength = random.randint(1, 10)  # 投げる強さ（1〜10）
    if strength == 6:
        reward = 10  # ゴール！
    else:
        reward = 0  # はずれ
    score += reward
    print(f"{i+1}回目：強さ {strength} → {reward}点")

print("合計スコア：", score)
```

> このように「強さ」と「結果（報酬）」を何度もためして、「ちょうどいい強さ（＝6）」を見つけていくのがAIのやり方！

## **1. 強化学習ってなに？**

### **1.1 今までのおさらい**

| 学習の種類 | 特徴（とくちょう） | 例 |
|------------|---------------------|-----|
| **分類**     | 「なにか？」を当てる       | 猫か犬かを見分ける |
| **回帰**     | 「どれくらい？」を予測     | 明日の気温を予測 |
| **強化学習** | 「どう動けばよい？」を学ぶ | ゲームで敵を倒す！ |

### **1.2 強化学習のイメージは… ゲーム！**

強化学習は、ゲームをプレイして上手くなっていく方法に似ています！

- 失敗してもいいからとにかく**やってみる**
- 「うまくいったとき」にごほうび（＝報酬）をもらえる
- 「何をすればごほうびをもらえるか」を少しずつ覚える！

## **2. 例：ゲームで敵を倒すには？**

たとえば、ロボットが敵に会ったとき…

| 行動         | 結果             | ごほうび（点数） |
|--------------|------------------|------------------|
| にげた        | ゲームオーバー      | 0点              |
| たたかった    | まけた             | 1点              |
| よけてから攻撃 | 勝った！            | 10点             |

最初はうまくいかなくても…

→「よけてから攻撃」がいちばん得点が高い！とわかると、  
→ それを**くり返すように学んでいく！**

## **3. 強化学習の3つのキーワード**

### **① エージェント（Agent）**
→ 行動する「ロボット」や「プレイヤー」のこと！

### **② 環境（Environment）**
→ ゲームのステージや、まわりの世界。

### **③ 報酬（Reward）**
→ うまくいったときにもらえる点数！

> 強化学習は、「どうすれば報酬がもらえるか？」を学ぶしかけ！

## **4. どうやって学ぶの？**

### **4.1 試行錯誤（しこうさくご）で覚える！**

- はじめは適当に動いてみる
- 「これはダメだ」「これはうまくいった」と体験をためる
- 少しずつ「いい動き」ができるようになる！

### 人間も同じ！

- 自転車に初めて乗ったとき、何度も転んだよね？
- でも「こうすればうまくいく！」を体で覚えた！

強化学習は、**人間の学び方といちばん似ている**んだよ！

## **5. アクティビティ：迷路ゲームをクリアせよ！**

### ゲームのねらい：
- 自分で動いてみて、「どうすればゴールできるか？」を体験する
- 強化学習の考え方を遊びながら理解する！

### 準備：
- 迷路（紙または簡単なボードゲーム）
- スタートとゴール、落とし穴などを用意

### ゲームの流れ：

1. **スタート地点にロボット（自分）を置く！**
2. **1歩ずつ進んでみる**
　→ どっちに行く？上？下？右？左？

3. **良いことがあったら得点！**
　- 宝箱に当たった → +5点
　- ゴールに着いた！→ +10点
　- 落とし穴に落ちた → -5点！

4. **繰り返して「一番得点が高くなるルート」を見つけよう！**

### ポイント！
- 最初は失敗してもOK！
- プレイをくり返すことで、「どの道がよかったか」を覚えていく！
- → これが「強化学習」！

## **6. Pythonでシンプルな強化学習のまねごと（応用編）**

※ 小学生〜中学生向けに簡単に紹介だけします。実装は大人向け。

```python
import random

# 行動の選び方
actions = ["にげる", "たたかう", "よけてから攻撃"]
rewards = {"にげる": 0, "たたかう": 1, "よけてから攻撃": 10}

# 10回ランダムにプレイ
for i in range(10):
    action = random.choice(actions)
    reward = rewards[action]
    print(f"{i+1}回目: {action} → {reward}点")
```

> こんなふうに、**いろんな行動をためして**、「どれが一番よかったかな？」を考えるのが強化学習です！
